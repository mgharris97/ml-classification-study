```python
# Scatter Plot 1


import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

BASE_DIR = Path().resolve().parent
DATA_PATH = BASE_DIR / "data" / "winequality-red-cleaned.csv"

df = pd.read_csv(DATA_PATH, sep=";")

colors = {"low": "red", "medium": "blue", "high": "green"}

for label, group in df.groupby("quality_label"):
    plt.scatter(group["alcohol"], group["sulphates"], 
                c=colors[label], label=label, alpha=0.5)

plt.xlabel("Alcohol")
plt.ylabel("Sulphates")
plt.title("Alcohol vs Sulphates by Quality Class")
plt.legend()
plt.show()
```
